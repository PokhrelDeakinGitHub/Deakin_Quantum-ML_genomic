```python
import plotly.graph_objects as go
import numpy as np

# Example setup (ensure you have actual data and iteration_list variable set up)
iteration_list = 50  # Define your iteration_list if not already defined
# Assuming QNN_Data and VQC_Data dictionaries are filled with relevant data

fig = go.Figure()

# Subplot 1: QNN Data
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=QNN_Data['Z'],
    mode='lines',
    name='ZFeatureMap',
    line=dict(color='purple')
))
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=QNN_Data['ZZ'],
    mode='lines',
    name='ZZFeatureMap'
))
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=QNN_Data['Pa'],
    mode='lines',
    name='PauliFeatureMap'
))

# Subplot 2: VQC Data
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=VQC_Data['Z'],
    mode='lines',
    name='ZFeatureMap (VQC)',
    line=dict(color='purple'),
    xaxis='x2',
    yaxis='y2'
))
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=VQC_Data['ZZ'],
    mode='lines',
    name='ZZFeatureMap (VQC)',
    xaxis='x2',
    yaxis='y2'
))
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=VQC_Data['Pa'],
    mode='lines',
    name='PauliFeatureMap (VQC)',
    xaxis='x2',
    yaxis='y2'
))

# Update layout for a dual subplot look
fig.update_layout(
    title="Objective Function Value VS Iteration",
    xaxis=dict(
        domain=[0, 0.45],
        title="Iteration",
        tickmode='array',
        tickvals=np.arange(0, iteration_list+5, 5),
        ticktext=[f"{i}" for i in range(0, iteration_list+5, 5)]
    ),
    yaxis=dict(
        title="Objective function value (QNN)",
        tickmode='array',
        tickvals=np.linspace(0.502, 0.486, 11),
        ticktext=[f"{y:.3f}" for y in np.linspace(0.502, 0.486, 11)]
    ),
    xaxis2=dict(
        domain=[0.55, 1],
        title="Iteration",
        tickmode='array',
        tickvals=np.arange(0, iteration_list+5, 5),
        ticktext=[f"{i}" for i in range(0, iteration_list+5, 5)]
    ),
    yaxis2=dict(
        title="Objective function value (VQC)",
        tickmode='array',
        tickvals=np.linspace(1.530, 1.02, 15),
        ticktext=[f"{y:.3f}" for y in np.linspace(1.530, 1.02, 15)],
        anchor="x2"
    ),
    width=1200,
    height=400
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Example setup
iteration_list = 50  # Number of iterations
# Generate example data if not defined
# QNN_Data and VQC_Data dictionaries should be filled with relevant data

fig = go.Figure()

# Subplot 1: QNN Data
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=QNN_Data['Z'],
    mode='lines',
    name='ZFeatureMap (QNN)',
    line=dict(color='purple')
))
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=QNN_Data['ZZ'],
    mode='lines',
    name='ZZFeatureMap (QNN)'
))
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=QNN_Data['Pa'],
    mode='lines',
    name='PauliFeatureMap (QNN)'
))

# Subplot 2: VQC Data, plotting ZFeatureMap on a secondary y-axis due to its wide range
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=VQC_Data['Z'],
    mode='lines',
    name='ZFeatureMap (VQC)',
    line=dict(color='purple'),
    xaxis='x2',
    yaxis='y3'  # Assign to the secondary y-axis for this subplot
))

# Plotting the other VQC features on the primary y-axis of subplot 2
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=VQC_Data['ZZ'],
    mode='lines',
    name='ZZFeatureMap (VQC)',
    xaxis='x2',
    yaxis='y2'
))
fig.add_trace(go.Scatter(
    x=list(range(0, iteration_list)),
    y=VQC_Data['Pa'],
    mode='lines',
    name='PauliFeatureMap (VQC)',
    xaxis='x2',
    yaxis='y2'
))

# Update layout with multiple y-axes configurations
fig.update_layout(
    title="Objective Function Value VS Iteration",
    xaxis=dict(
        domain=[0, 0.45],
        title="Iteration (QNN)",
        tickmode='array',
        tickvals=np.arange(0, iteration_list + 5, 5),
        ticktext=[f"{i}" for i in range(0, iteration_list + 5, 5)]
    ),
    yaxis=dict(
        title="Objective function value (QNN)"
    ),
    xaxis2=dict(
        domain=[0.55, 1],
        title="Iteration (VQC)",
        tickmode='array',
        tickvals=np.arange(0, iteration_list + 5, 5),
        ticktext=[f"{i}" for i in range(0, iteration_list + 5, 5)]
    ),
    yaxis2=dict(
        title="Objective function value (VQC ZZ/Pa)",
        anchor="x2",
        side='left'
    ),
    yaxis3=dict(
        title="Objective function value (VQC Z)",
        anchor="free",
        overlaying="y2",
        side='right',
        position=0.95  # Adjust this as needed to position the y-axis correctly
    ),
    width=1200,
    height=400
)

fig.show()
```
